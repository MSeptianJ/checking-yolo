## CLI Usage

```python
!yolo task=detect mode=predict model=yolov8n.pt source="/content/sample.jpg"
```

```python
filename = "sample.jpg"
path = "/content/{}".format(filename)
resultPath = "/content/runs/detect/predict/{}".format(filename)

img_before = cv2.imread(path)
img_after = cv2.imread(resultPath)

img_before = cv2.cvtColor(img_before, cv2.COLOR_BGR2RGB)
img_after = cv2.cvtColor(img_after, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(211)
plt.title("Before")
plt.imshow(img_before)
plt.axis("off")

plt.subplot(212)
plt.title("After")
plt.imshow(img_after)
plt.axis("off")

plt.show()
```

## Python Usage

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")

results = model(path)
res_plotted = results[0].plot()
end = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(211)
plt.title("Before")
plt.imshow(img_before)
plt.axis("off")

plt.subplot(212)
plt.title("After")
plt.imshow(end)
plt.axis("off")

plt.show()
```
